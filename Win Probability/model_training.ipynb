```python
# Compute cs/min for each role (except for supports as they don't farm)
def compute_cs_min(all_data):
    all_data['BLUE TOP CS/MIN'] = round(all_data['BLUE TOP CS'] / (all_data['TIME STAMP']/60000),1)
    all_data['BLUE JUNGLE CS/MIN'] = round(all_data['BLUE JUNGLE CS'] / (all_data['TIME STAMP']/60000),1)
    all_data['BLUE MID CS/MIN'] = round(all_data['BLUE MID CS'] / (all_data['TIME STAMP']/60000),1)
    all_data['BLUE ADC CS/MIN'] = round(all_data['BLUE ADC CS'] / (all_data['TIME STAMP']/60000),1)


    all_data['RED TOP CS/MIN'] = round(all_data['RED TOP CS'] / (all_data['TIME STAMP']/60000),1)
    all_data['RED JUNGLE CS/MIN'] = round(all_data['RED JUNGLE CS'] / (all_data['TIME STAMP']/60000),1)
    all_data['RED MID CS/MIN'] = round(all_data['RED MID CS'] / (all_data['TIME STAMP']/60000),1)
    all_data['RED ADC CS/MIN'] = round(all_data['RED ADC CS'] / (all_data['TIME STAMP']/60000),1)

    
    return all_data
```

```python
def mapping_turrets(turret):
    # Check for top turrets
    if turret == 'Turret_T2_L_03_A':
        return 'TOP_LANE OUTER_TURRET'
    elif turret == 'Turret_T2_L_02_A':
        return 'TOP_LANE INNER_TURRET'
    elif turret == 'Turret_T2_L_01_A':
        return 'TOP_LANE BASE_TURRET'
    
    # Check for mid turrets
    elif turret == 'Turret_T2_C_05_A':
        return 'MID_LANE OUTER_TURRET'
    elif turret == 'Turret_T2_C_04_A':
        return 'MID_LANE INNER_TURRET'
    elif turret == 'Turret_T2_C_03_A':
        return 'MID_LANE BASE_TURRET'
    elif turret == 'Turret_T2_C_02_A' or turret == 'Turret_T2_C_01_A':
        return 'MID_LANE NEXUS_TURRET'
    
    # Check for bot turrets
    elif turret == 'Turret_T2_R_03_A':
        return 'BOT_LANE OUTER_TURRET'
    elif turret == 'Turret_T2_R_02_A':
        return 'BOT_LANE INNER_TURRET'
    elif turret == 'Turret_T2_R_01_A':
        return 'BOT_LANE BASE_TURRET'
    
    
```

```python
def mapping_inhibs(inhib):
    if inhib == 'Barracks_T2_L1':
        return 'TOP'
    elif inhib == 'Barracks_T2_C1':
        return 'MID'
    elif inhib == 'Barracks_T2_R1':
        return 'BOT'
```
